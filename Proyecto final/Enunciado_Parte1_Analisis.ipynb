> 1. El problema corresponde a un problema de clasificación supervisada multiclase. Es supervisada porque se cuenta con datos históricos de tickets que ya fueron priorizados manualmente, por lo tanto existe una variable objetivo conocida. Es multiclase porque el modelo debe asignar cada ticket a una de cuatro categorías posibles de prioridad: Baja, Media, Alta o Crítica.
No corresponde a un problema de regresión, ya que no se busca predecir un valor numérico continuo, como un monto o un tiempo estimado de resolución. Tampoco corresponde a clustering, porque no se busca descubrir grupos desconocidos dentro de los datos, sino aprender a partir de clases previamente definidas. Podría existir un enfoque de ranking para ordenar tickets según urgencia, pero el enunciado nos pide explícitamente predecir una categoría de prioridad, por lo que el enfoque más adecuado es clasificación.
El problema se relaciona con procesamiento de lenguaje natural, ya que una parte importante de la información viene del texto escrito por los clientes en los tickets. La literatura sobre clasificación automática de tickets de soporte señala que automatizar este proceso puede mejorar el desempeño del área de atención al cliente y reducir tiempos de resolución. (señalo la pagína que encontre con información con respecto a este tema arvix.org) .

> 2. El negocio solicita resolver este problema porque ChaucherApp recibe una gran cantidad de tickets de soporte y necesita asignar sus recursos de manera eficiente. No todos los tickets tienen la misma urgencia ni el mismo impacto, por ejemplo, un problema de consulta general no debería recibir la misma prioridad que un ticket asociado a una falla de pago, bloqueo de cuenta o fraude.
La correcta priorización permite que los tickets más urgentes sean atendidos antes, reduciendo riesgos operacionales, pérdidas de confianza y deterioro en la experiencia del cliente. Sobre todo esto es importante porque una mala atención al cliente puede afectar directamente la reputación de la empresa, la retención de usuarios y la percepción de seguridad del servicio.

> 3. El negocio solicita un modelo predictivo porque los tickets de soporte están escritos en lenguaje natural, por lo que pueden expresar el mismo problema de muchas formas distintas. Por ejemplo, un usuario podría escribir “no puedo pagar”, “mi tarjeta no funciona”, “la app rechaza la compra” o “me aparece error al comprar”, y todas esas frases podrían referirse a problemas similares, aunque no usen exactamente las mismas palabras.
Un sistema basado solo en palabras clave sería rígido, difícil de mantener y propenso a errores. Tendría que anticipar manualmente todas las formas posibles en que los clientes describen un problema. 
En cambio, un modelo de ML puede aprender patrones desde los datos históricos, considerando combinaciones de palabras, contexto y atributos adicionales. Esto permite una clasificación más flexible y adaptable. 

> 4. Las métricas relevantes que yo buscaría optimizar serían:
 >- Tiempo de primera respuesta por prioridad: mide cuánto tarda el equipo en dar la primera respuesta a un ticket desde que fue ingresado. Es especialmente relevante para tickets de prioridad Alta o Crítica, ya que estos deberían ser atendidos rápidamente. Sobre todo porque las personas no son pacientes entonces reducir el tiempo de respuesta permitira que tengan una mejor percepción de la emrpesa.
> - Tiempo promedio o mediano de resolución: mide cuánto tarda la empresa en resolver completamente los tickets. Si el modelo prioriza correctamente, debería ayudar a reducir el tiempo de resolución de los tickets más urgentes.
> - Tiempo de resolución de ticket por clase de prioridad: midiendo qué porcentaje de tickets se resuelve dentro del tiempo comprometido para cada nivel de prioridad. Esto es importante porque no todas las prioridades deberían tener el mismo plazo de atención. 
> - Satisfacción del cliente: permite evaluar si la automatización realmente mejora la experiencia del usuario. Un modelo podría ser técnicamente bueno, pero si genera malas asignaciones o retrasa tickets urgentes, la satisfacción podría disminuir.

> 5. Como este es un problema de clasificación multiclase, las métricas más importantes son accuracy, precision, recall y F1-score. Para interpretarlas de manera simple, se puede pensar primero en un problema binario, una predicción puede estar Correcta o Incorrecta. Luego, en el caso multiclase, estas métricas se calculan para cada clase usando una lógica uno contra los demas, por ejemplo, “Crítica” versus “No Crítica”.
> - Accuracy: mide la proporción total de tickets correctamente clasificados. Es decir, de todos los tickets evaluados, cuántos fueron asignados a la prioridad correcta. Scikit-learn define accuracy como la fracción de predicciones correctas sobre el total de muestras.
> - Precision: mide, para una clase específica, qué proporción de los tickets que el modelo predijo como esa clase realmente pertenecían a ella. Por ejemplo, si el modelo predice 100 tickets como Críticos y 80 efectivamente eran Críticos, la precision de la clase Crítica sería 80%. Esta métrica ayuda a controlar los falsos positivos.
> - Recall: mide, para una clase específica, qué proporción de los tickets que realmente pertenecían a esa clase fueron correctamente detectados por el modelo. Por ejemplo, si existían 100 tickets realmente Críticos y el modelo detectó 85, el recall de la clase Crítica sería 85%. Esta métrica es muy importante cuando el costo de no detectar una clase es alto.
> - F1-score: combina precision y recall en una sola métrica mediante un promedio armónico. Es útil cuando se busca equilibrio entre no sobreclasificar una clase y no dejar casos importantes sin detectar. 
> - MSE: mide el promedio de los errores al cuadrado entre el valor real y el valor predicho. Se utiliza cuando se quiere predecir un número, por ejemplo, el tiempo de resolución de un ticket en horas. Mientras más bajo sea el MSE, mejor es el modelo. 
> - RMSE: corresponde a la raíz cuadrada del MSE. Tiene la ventaja de quedar expresado en la misma unidad de la variable que se predice. Por ejemplo, si se estuviera prediciendo el tiempo de resolución en horas, el RMSE también estaría expresado en horas.
> - R²: mide qué proporción de la variabilidad de una variable numérica es explicada por el modelo. Un R² cercano a 1 indica que el modelo explica bien los datos, mientras que un valor cercano a 0 indica bajo poder explicativo. Esta métrica se utiliza en regresión, por ejemplo, para evaluar qué tan bien se predice el tiempo de atención de un ticket.
> En este problema, métricas de regresión como MSE, RMSE o R² no son las más adecuadas, porque la variable objetivo no es numérica continua, sino una etiqueta categórica de prioridad.

> 6. Dado que las clases representan distintos niveles de urgencia, no todas deberían evaluarse con el mismo criterio principal:
> - Para la clase crítica la metrica más importante es Recall, pues lo más importante es no dejar pasar tickets realmente críticos. Un falso negativo en esta clase puede significar que un caso urgente sea tratado como menos importante.
> - Para la clase alta la metrica más importante es Recall o F1-score, pues también es importante detectar la mayor cantidad posible de tickets urgentes, aunque manteniendo cierto equilibrio para no saturar al equipo con demasiados falsos positivos.
> - Para la clase media la metrica más importante es F1-score, pues en esta clase conviene equilibrar precision y recall, ya que no es tan crítica como Alta o Crítica, pero igualmente debe ser clasificada correctamente para ordenar la carga operativa.
> - Para la clase baja la metrica más importante es Precision, pues es importante que los tickets clasificados como Baja realmente sean de baja prioridad. Si un ticket urgente se clasifica erróneamente como Baja, podría quedar postergado injustificadamente.

> 7. La agregación macro calcula la métrica por separado para cada clase y luego obtiene un promedio simple, dando el mismo peso a todas las clases. Esto es útil cuando se quiere evaluar si el modelo funciona bien en todas las categorías, incluso en aquellas con menos datos.
La agregación weighted también calcula la métrica por clase, pero luego promedia ponderando por la cantidad de ejemplos reales de cada clase. Esto refleja mejor el desempeño global cuando hay clases más frecuentes, pero puede ocultar malos resultados en clases minoritarias.
Para este caso, utilizaría macro F1-score como métrica principal, porque en soporte al cliente las clases minoritarias pueden ser justamente las más importantes, como los tickets Críticos. Si solo se usara weighted F1-score, el desempeño podría verse artificialmente alto si el modelo clasifica bien muchos tickets de Baja o Media, pero falla en los Críticos.
De todas maneras, también usaría weighted F1-score como métrica complementaria, porque permite observar el rendimiento general del modelo considerando la distribución real de tickets.

```python
############ 3 ##########
from plotly.subplots import make_subplots
import plotly.express as px

variables_histogramas = [
    "N_Caracteres_Ticket",
    "Tiempo_de_Resolución_Horas",
    "Usuario-Antiguedad_Cuenta_Dias",
    "Canal_Ticket",
    "Categoría_Problema",
    "Usuario-Tipo_de_Cuenta"
]

# Verificar si todas las columnas existen
columnas_faltantes = [col for col in variables_histogramas if col not in df_tickets.columns]

if columnas_faltantes:
    print("Columnas que no existen en el dataframe:")
    print(columnas_faltantes)
else:
    print("Todas las columnas existen correctamente.")


fig = make_subplots(
    rows=3,
    cols=2,
    subplot_titles=variables_histogramas
)

for i, col in enumerate(variables_histogramas):
    hist = px.histogram(
        df_tickets,
        x=col
    )
    
    fila = i // 2 + 1
    columna = i % 2 + 1
    
    fig.add_trace(hist.data[0], row=fila, col=columna)

fig.update_layout(
    title="Distribución de características numéricas y categóricas",
    height=900,
    showlegend=False
)

fig.show()
```

```python
######### 4 ##########
columnas_numericas = df_tickets.select_dtypes(include=["int64", "float64"]).columns

matriz_corr = df_tickets[columnas_numericas].corr()

matriz_corr

fig = px.imshow(
    matriz_corr,
    text_auto=True,
    title="Matriz de correlación entre variables numéricas"
)

fig.show()
```

```python
########### 5 ################
#Crear bins
df_tickets["Bin_N_Caracteres_Ticket"] = pd.qcut(
    df_tickets["N_Caracteres_Ticket"],
    q=3,
    labels=["Bajo", "Medio", "Alto"]
)

df_tickets["Bin_Tiempo_de_Resolución_Horas"] = pd.qcut(
    df_tickets["Tiempo_de_Resolución_Horas"],
    q=3,
    labels=["Bajo", "Medio", "Alto"]
)

df_tickets["Bin_Usuario_Antiguedad_Cuenta_Dias"] = pd.qcut(
    df_tickets["Usuario-Antiguedad_Cuenta_Dias"],
    q=3,
    labels=["Bajo", "Medio", "Alto"]
)

df_tickets[[
    "Bin_N_Caracteres_Ticket",
    "Bin_Tiempo_de_Resolución_Horas",
    "Bin_Usuario_Antiguedad_Cuenta_Dias"
]].head()

# Crear tablas cruzadas normalizadas por fila
ct_categoria = pd.crosstab(
    df_tickets["Categoría_Problema"],
    df_tickets["Nivel_Prioridad"],
    normalize="index"
)

ct_canal = pd.crosstab(
    df_tickets["Canal_Ticket"],
    df_tickets["Nivel_Prioridad"],
    normalize="index"
)

ct_tipo_cuenta = pd.crosstab(
    df_tickets["Usuario-Tipo_de_Cuenta"],
    df_tickets["Nivel_Prioridad"],
    normalize="index"
)

ct_caracteres = pd.crosstab(
    df_tickets["Bin_N_Caracteres_Ticket"],
    df_tickets["Nivel_Prioridad"],
    normalize="index"
)

ct_tiempo = pd.crosstab(
    df_tickets["Bin_Tiempo_de_Resolución_Horas"],
    df_tickets["Nivel_Prioridad"],
    normalize="index"
)

ct_antiguedad = pd.crosstab(
    df_tickets["Bin_Usuario_Antiguedad_Cuenta_Dias"],
    df_tickets["Nivel_Prioridad"],
    normalize="index"
)

#Graficar contingencias en subplots con imshow
tablas = [
    ct_categoria,
    ct_canal,
    ct_tipo_cuenta,
    ct_caracteres,
    ct_tiempo,
    ct_antiguedad
]

titulos = [
    "Categoría del Problema vs Nivel de Prioridad",
    "Canal del Ticket vs Nivel de Prioridad",
    "Tipo de Cuenta vs Nivel de Prioridad",
    "N° Caracteres Ticket vs Nivel de Prioridad",
    "Tiempo de Resolución vs Nivel de Prioridad",
    "Antigüedad Cuenta vs Nivel de Prioridad"
]

fig = make_subplots(
    rows=3,
    cols=2,
    subplot_titles=titulos
)

for i, tabla in enumerate(tablas):
    row = i // 2 + 1
    col = i % 2 + 1
    
    heatmap = go.Heatmap(
        z=tabla.values,
        x=tabla.columns,
        y=tabla.index,
        text=tabla.round(2).values,
        texttemplate="%{text}",
        showscale=False
    )
    
    fig.add_trace(heatmap, row=row, col=col)

fig.update_layout(
    title="Tablas de contingencia normalizadas por fila",
    height=1000
)

fig.show()
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
from pathlib import Path

from umap import UMAP

```

```python
#Leer el archivo embeddings.parquet
ruta_embeddings = Path("data/embeddings.parquet")

df_embeddings = pd.read_parquet(ruta_embeddings)

df_embeddings.head()
```

```python
df_embeddings.info()
df_embeddings.columns
```

```python
#Alinear embeddings con df_tickets
# Revisar tamaños
print("Tickets:", df_tickets.shape)
print("Embeddings:", df_embeddings.shape)

# Revisar primeros índices de embeddings
df_embeddings.index[:5]
```

```python
# Alinear embeddings con el orden de df_tickets usando Id_Ticket
df_embeddings_ordenado = df_embeddings.loc[df_tickets["Id_Ticket"]]

# Verificar que quedaron en el mismo orden
print(df_embeddings_ordenado.shape)
print(df_tickets["Id_Ticket"].head())
print(df_embeddings_ordenado.index[:5])
```

```python
#Crear matriz de embeddings

X_embeddings = df_embeddings_ordenado.values

print("Shape matriz embeddings:", X_embeddings.shape)
```

```python
#Aplicamos UMAP
umap_model = UMAP(
    n_components=2,
    random_state=42,
    n_neighbors=15,
    min_dist=0.1
)

proyecciones = umap_model.fit_transform(X_embeddings)

proyecciones[:5]
```

```python
# Nuevas dimensiones a df_tickets
df_tickets["x"] = proyecciones[:, 0]
df_tickets["y"] = proyecciones[:, 1]

df_tickets[["Id_Ticket", "x", "y", "Nivel_Prioridad", "Categoría_Problema"]].head()
```

```python
#Scatter plot por Nivel_Prioridad
import textwrap


def wrap_text(text, width=40):
    return "<br>".join(textwrap.wrap(text, width=width))


df_tickets["Contenido_Ticket_Wrapped"] = df_tickets["Contenido_Ticket"].apply(lambda x: wrap_text(x, 80))

fig_proj = px.scatter(
    df_tickets,
    x="x",
    y="y",
    color="Nivel_Prioridad",
    title="Proyecciones de los embeddings respecto a clase",
    hover_data=[
        "Canal_Ticket",
        "N_Caracteres_Ticket",
        "Categoría_Problema",
        "Usuario-Tipo_de_Cuenta",
        "Usuario-Antiguedad_Cuenta_Dias",
        "Contenido_Ticket_Wrapped",
    ],
)
fig_proj.show()
```

```python
# Scatter plot por Categoría_Problema
fig_proj_categoria = px.scatter(
    df_tickets,
    x="x",
    y="y",
    color="Categoría_Problema",
    title="Proyecciones de los embeddings respecto a Categoría del Problema",
    hover_data=[
        "Nivel_Prioridad",
        "Canal_Ticket",
        "N_Caracteres_Ticket",
        "Usuario-Tipo_de_Cuenta",
        "Usuario-Antiguedad_Cuenta_Dias",
        "Contenido_Ticket_Wrapped",
    ],
)

fig_proj_categoria.show()
```

> 1. Un embedding es una representación numérica de un dato, en este caso texto, mediante un vector de muchas dimensiones. En vez de trabajar directamente con palabras, el modelo transforma el contenido textual del ticket en una secuencia de números que intenta capturar su significado, contexto y similitud semántica. Para este proyecto, los embeddings fueron calculados con el modelo gemini-embedding-001, que recibe texto como entrada y entrega embeddings de texto como salida. Según la documentación de Google AI, este modelo soporta dimensiones de salida flexibles entre 128 y 3072, con tamaños recomendados de 768, 1536 y 3072 dimensiones. En nuestro caso, el archivo embeddings.parquet contiene 1024 dimensiones para cada ticket, desde embedding_dim_1 hasta embedding_dim_1024.

> 2. Los embeddings pueden ser mejores que analizar palabras individuales porque no solo consideran la presencia de términos aislados, sino también el significado general del texto. Por ejemplo, frases como “no puedo pagar”, “la compra fue rechazada” o “mi tarjeta no funciona” pueden describir problemas parecidos aunque usen palabras distintas.\\
Un enfoque basado solo en palabras individuales puede perder estas relaciones semánticas. En cambio, los embeddings permiten que textos similares queden más cercanos entre sí, incluso si no comparten exactamente las mismas palabras. Esto puede ayudar a que el modelo predictivo detecte patrones más robustos en los tickets.

> 3. Una desventaja es que los embeddings son menos interpretables. Cuando se trabaja con palabras directamente, es más fácil explicar que cierta palabra o frase influyó en la predicción. En cambio, un embedding está compuesto por muchas dimensiones numéricas, por lo que no es evidente qué representa cada dimensión individual. Además, usar embeddings puede aumentar la complejidad computacional del modelo. En este caso, cada ticket tiene 1024 variables provenientes del embedding, lo que puede hacer que el entrenamiento sea más costoso y más difícil de interpretar.

> 4. Sí, existen distintos tipos de embeddings. En este proyecto se utilizan embeddings de texto, que representan el significado del asunto y contenido del ticket. Sin embargo, también existen embeddings para imágenes, audio, video, usuarios, productos o documentos.
